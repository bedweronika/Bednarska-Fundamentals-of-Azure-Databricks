```python
dbutils.help()
```

* CREDENTIALS - allows to show user roles
* DATA - allows to understand a dataset as on date , to calculate some statistics on a data frame and gives a summary
* FS - gives a method to interact with the databticks file system DBFS
* JOBS - interact wit hthe workflow jobs
* LIBRARY - allows to locally compile an application 
* META - allows to hook into the compiler
* NOTEBOOK - allows to run  chain of notebook 
* PREVIEW - for hidden utilites under preview
* SECRETS - access to secure credentials within the notebook
* WIDGETS - creates notebook parameters

Tne notebook utilities has only 2 methods: run and exit. To check these methods I will create a new notebook: CHILD-NOTEBOOK


```python
# use child notebook
dbutils.notebook.run("./child-notebook-utilities", 10)
```

```python
# now we call the child notebook with parameter for message

dbutils.notebook.run("./child-notebook-utilities", 10, {"msg": " call from parent notebook"})
```
