```python
dbutils.widgets.help("text")
```

```python
############ THIS CREATE A TEXT GAP on the top of the notebook

dbutils.widgets.text("msg", "", "Your input parameter")
# first argument is the name of the widget
# second argument is the default value
# third argument is the label
```

```python
p_msg = dbutils.widgets.get("msg")
```

```python
print("I am a child notebook and recived: " + p_msg)
```
